# Info Data

### Melihat Info Dataset
Karena dataset yang kita gunakan data type-nya masih berbentuk object, maka akan kita lakukan transform tipe data ke dalam bentuk float64.

```python
df.info()
```

### Mengubah Tipe Data
Kode di bawah akan mengubah tipe data pada tabel kecuali kolom pertama yang dibiarkan agar tetap object dikarenakan mengandung baris string.

```python
cols_to_transform = list(df.columns)[1:]

def transform_columns(df, cols):
  for col in cols:
    df[col] = pd.to_numeric(df[col].str.replace(',', ''), errors='coerce')

  return df

df = transform_columns(df.copy(), cols_to_transform)
print(df)
```

Dari informasi di atas, terlihat bahwa tipe data untuk kolom selain kolom 'PDB Lapangan Usaha' sudah terganti menjadi float64.

## Kesimpulan Data

### Interpretasi:

- **Ukuran titik**: Ukuran titik menunjukkan kontribusi lapangan usaha terhadap PDB. Semakin besar titik, semakin besar kontribusinya.
- **Posisi titik**: Posisi titik menunjukkan kinerja lapangan usaha. Titik yang berada di atas garis tren menunjukkan kinerja yang lebih baik dibandingkan rata-rata, sedangkan titik yang berada di bawah garis tren menunjukkan kinerja yang kurang baik.
- **Perubahan tren**: Perubahan tren menunjukkan perubahan arah pergerakan PDB di setiap lapangan usaha.

### Contoh:

- **Pertanian, Peternakan, Kehutanan, Perikanan**: Kontribusi sektor ini terhadap PDB relatif kecil dibandingkan sektor lain. Kinerja sektor ini relatif stabil dengan tren yang sedikit menurun.
- **Pertambangan dan Penggalian**: Kontribusi sektor ini terhadap PDB cukup besar. Kinerja sektor ini fluktuatif dengan tren yang sedikit menurun.
- **Industri Pengolahan**: Kontribusi sektor ini terhadap PDB paling besar. Kinerja sektor ini relatif stabil dengan tren yang sedikit meningkat.

### Kesimpulan:

Gambar tersebut menunjukkan bahwa PDB di Indonesia didominasi oleh sektor **Industri Pengolahan**. Sektor ini juga menunjukkan kinerja yang paling stabil dibandingkan sektor lain.
